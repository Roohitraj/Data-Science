```python
#7
correlation = dfs["no_of_trainings"].corr(dfs["avg_training_score"])
print(f"Correlation between No. of Trainings and Avg Training Score: {correlation:.2f}")
if correlation > 0.7:
    print(f"Correlation = {correlation:.2f} → Strong Positive Correlation (More training → Higher performance)")
elif correlation > 0.3:
    print(f"Correlation = {correlation:.2f} → Moderate Positive Correlation (More training may lead to better performance)")
elif correlation > 0:
    print(f"Correlation = {correlation:.2f} → Weak Positive Correlation (Training has a slight impact on performance)")
elif correlation < -0.7:
    print(f"Correlation = {correlation:.2f} → Strong Negative Correlation (More training → Lower performance)")
elif correlation < -0.3:
    print(f"Correlation = {correlation:.2f} → Moderate Negative Correlation (More training may slightly decrease performance)")
elif correlation < 0:
    print(f"Correlation = {correlation:.2f} → Weak Negative Correlation (Slight negative effect of training on performance)")
else:
    print(f"Correlation = {correlation:.2f} → No Correlation (Training count does not significantly impact performance) ❓")
```
